```python
import numpy as np
from numpy import linalg
from scipy import stats
import matplotlib.pyplot as plt
from skimage import io, data, color, feature, filters, exposure, transform, segmentation, measure
from skimage.future import graph
%matplotlib inline
%load_ext autoreload
%autoreload 2
plt.gray()
```

```python
def size(x=15, y=15):
    plt.figure(figsize=(x,y))
    
def bounds(xlo, xhi, ylo, yhi):
    plt.axes().set_xlim(xlo, xhi)
    plt.axes().set_ylim(ylo, yhi)
    
def imgbounds(img):
    bounds(0, img.shape[1], img.shape[0], 0)
    
def show(img, x=15, y=15):
    size(x, y)
    imgbounds(img)
    plt.imshow(img)
```

```python
img = io.imread('pics/0_0.jpg')
gimg = color.rgb2gray(img)
show(img)
```

```python
edges = feature.canny(gimg)
show(edges)
```

```python
size()
plt.imshow(edges * 0)
plt.axes().set_xlim((0,gimg.shape[1]))
plt.axes().set_ylim((gimg.shape[0],0))
lines = transform.probabilistic_hough_line(edges, threshold=50, line_length=50, line_gap=3)
print(len(lines), ' lines')
for line in lines:
    p0, p1 = line
    plt.plot((p0[0], p1[0]), (p0[1], p1[1]))
```

```python
some_lines = lines[:5]
def intersection(lines):
    points = np.array([pair[0] for pair in lines])
    norms = []
    for line in lines:
        x1, x2 = line
        d = np.array(x2) - np.array(x1)
        unit = d/linalg.norm(d)
        norms.append(np.atleast_2d(np.dot([[0,-1],[1,0]], unit)))
    left = np.array([np.dot(norm.T, norm) for norm in norms]).sum(axis=0)
    right = np.array([np.dot(np.dot(norms[i].T, norms[i]), points[i]) for i in range(len(norms))]).sum(axis=0)
    return np.dot(linalg.inv(left), right)
```

```python
x = intersection(some_lines)
imgbounds(img)
for line in some_lines:
    (x1,y1), (x2,y2) = line
    plt.plot([x1,x2],[y1,y2])
plt.scatter(x[0], x[1], marker='+')
```
