```python
def size(x=15, y=15):
    plt.figure(figsize=(x,y))
    
def bounds(xlo, xhi, ylo, yhi):
    plt.axes().set_xlim(xlo, xhi)
    plt.axes().set_ylim(ylo, yhi)
    
def imgbounds(img):
    bounds(0, img.shape[1], img.shape[0], 0)
    
def show(img, x=15, y=15):
    size(x, y)
    imgbounds(img)
    plt.imshow(img)
```

```python
class BestIntersection:
    
    def points_norms(self, lines):
        points = np.array([pair[0] for pair in lines])
        norms = []
        for line in lines:
            x1, x2 = line
            d = np.array(x2) - np.array(x1)
            unit = d/linalg.norm(d)
            norms.append(np.atleast_2d(np.dot([[0,-1],[1,0]], unit)))
        return points, norms
    
    def estimate(self, lines):
        n = len(lines)
        points, norms = self.points_norms(lines)
        left = np.array([np.dot(norm.T, norm) for norm in norms]).sum(axis=0)
        right = np.array([np.dot(np.dot(norms[i].T, norms[i]), points[i]) for i in range(n)]).sum(axis=0)
        self.x = np.dot(linalg.inv(left), right)
    
    def residuals(self, lines):
        n = len(lines)
        points, norms = self.points_norms(lines)
        res = []
        for i in range(n):
            part1 = (x-points[i]).T
            part2 = np.dot(norms[i].T,norms[i])
            part3 = x-points[i]
            res.append(np.dot(np.dot(part1,part2),part3))
        return res
```

```python
road_lines = []
for line in lines:
    (x1,y1), (x2,y2) = line
    if road[y1,x1] and road[y2,x2]:
        road_lines.append(line)
    
point, inliers = measure.ransac(np.array(road_lines), BestIntersection, min_samples=3, residual_threshold=10.0)
```
